```python
# Looking at a Star Wars Survey
```

```python
import pandas as pd
import numpy as np

starwars = pd.read_csv("StarWars.csv", encoding = "latin1")

starwars.head(10)
```

```python
starwars.columns
```

```python
#remove the null RespondentID's

star_wars = starwars[pd.notnull(starwars["RespondentID"])]

print(star_wars.head(5))
```

```python
yes_no_map = {
    "Yes": True,
    "No": False
}

star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no_map)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no_map)

star_wars.head(5)
```

```python
#rename the column values
movie_map = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False
}


for something in star_wars.columns[3:9]:
    star_wars[something] = star_wars[something].map(movie_map)
    
#rename the columns
movie_column_map = {
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6"    
}

star_wars = star_wars.rename(columns = movie_column_map)

star_wars.head(5)
```

```python
#convert rankings into float type
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

#rename ranking columns
rank_map = {
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': "rank_of_1",
    "Unnamed: 10": "rank_of_2",
    "Unnamed: 11": "rank_of_3",
    "Unnamed: 12": "rank_of_4",
    "Unnamed: 13": "rank_of_5",
    "Unnamed: 14": "rank_of_6"
}


star_wars = star_wars.rename(columns = rank_map)

star_wars[star_wars.columns[8:16]].head(5)
```

```python
#Finding the highest ranked movies

rank_means = star_wars[star_wars.columns[9:15]].mean(axis = 0)

import matplotlib.pyplot as plt
%matplotlib inline

#add figure and axes
fig, ax = plt.subplots()

#assign a bar graph to an axes
ax.bar([1,2,3,4,5,6], rank_means)
ax.set_title("SW Movies Rankings, Lower is Better")
#ax.set_xticklabels(rank_means.index)
plt.show()

#lower average = higher ranking!
```

# Recap

So far, we've read in the Star Wars survey data, removed NaN ID entries, and renamed a few columns (and their data) for a more intelligible dataset. Then we found the average rankings of all six Star Wars movies and plotted them against one another.
We saw that the highest ranked movie was Episode 5: The Empire Strikes Back and that the original trilogy is ranked higher than the prequel trilogy across the board. What I found most surprising is that Episode 3: Revenge of the Sith was ranked the lowest among all six movies when it was clearly superior to Episode 2 (Episode 3 is my favorite of the saga).

```python
#Finding the most watched movie

rank_sum = star_wars[star_wars.columns[3:9]].sum()

fig, ax2 = plt.subplots()

ax2.bar([1,2,3,4,5,6], rank_sum)
ax2.set_title("Watch Counts of Each SW Movie")

plt.show()
```

The most noticeable result is that Episode 5, the __highest ranked__ movie, has been watched by the __most people__ while Episode 3, the __lowest ranked__ movie, has been watched the __least number of people__.

Voters may decide to rank the movies they haven't seen lower which would throw off the average rankings. Assuming people ranked the movies they watched first then arbitrarily ranked the ones they did not watch, we can expect Episode 3 to end up at the bottom.

# Exploring the Data by Binary Segments

We can separate the data based on whether the respondents considered themselves Star Wars or not.

```python
#separate into fans and non-fans
fans = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == True]
heathens = star_wars[star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] == False]

#Finding the highest ranked movies
rank_means_fans = fans[fans.columns[9:15]].mean(axis = 0)
rank_means_heathens = heathens[heathens.columns[9:15]].mean(axis = 0)

#Finding the watch counts
rank_sum_fans = fans[fans.columns[3:9]].sum()
rank_sum_heathens = heathens[heathens.columns[3:9]].sum()

#Figure setup
ind = [1, 2, 3, 4, 5, 6]
width = 0.35
fig = plt.figure()
ax = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

#Putting data on figure
ax.bar(ind, rank_means_fans, width, color = "r")
ax.bar([x + width for x in ind], rank_means_heathens, width, color = "black")
ax.set_title("Movie Rankings (Fans & Non-Fans), Fans = Red")

ax2.bar(ind, rank_sum_fans, width, color = "r")
ax2.bar([x + width for x in ind], rank_sum_heathens, width, color = "black")
ax2.set_title("Watch Counts (Fans & Non-Fans)")

plt.tight_layout()
plt.show()
```

## Fans vs. Non-Fans
For the movie rankings, self-proclaimed fans rank the original trilogy higher than non-fans while ranking the prequel trilogy lower than non-fans. Additionally, the order from favorite to least favorite for fans is 5, 6, 4, 1, 2, 3 while non-fans ranked the movies 5, 1, 6, 2, 4, 3. Non-fans ranked Episode 1 quite favorably overall and generally disliked Episode 4. 
If pressed for conjecture, I would say fans may be blinded by nostalgia or were caught up in the hate for the prequels which clouded their judgements.

On the watch counts side, we see that fans have watched dramatically more of all the movies than non-fans which is hardly surprising. The order from most to least watched is 5, 6, 1, 4, 2, 3 for fans and 5, 6, 1, 4, 2, 3 for non-fans. Interestingly enough, the order is the same for both groups! 

Perhaps both sides are not so dissimilar...
